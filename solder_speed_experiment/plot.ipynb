```python
X = np.array([1.25, 1.3, 1.4, 1.5, 1.75, 2, 2.25])
Y = np.array([96, 120, 210, 240, 300, 408, 552])
plt.plot(X, Y, '.')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X.reshape(-1, 1), Y)
```

```python
print(model.coef_, model.intercept_)
```

```python
x1, y1 = 1.2, model.predict([[1.2]])
x2, y2 = 2.4, model.predict([[2.4]])
print(x1, y1)
print(x2, y2)
```
