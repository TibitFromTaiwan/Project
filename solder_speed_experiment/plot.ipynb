```python
import matplotlib.pyplot as plt
import csv
```

```python
with open('record0429.csv', newline='') as csvfile:

    rows = csv.DictReader(csvfile)
    X = []
    Y = []
    for row in rows:
        print(row)
        X.append(row['Speed'])
        Y.append(row['Length Per Second'])
```

```python
print(X)
print(Y)
```

```python
plt.plot(X, Y)
```
